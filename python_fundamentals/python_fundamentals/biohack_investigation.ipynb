<b>üí° BUILT-IN DATA TYPES:</b><br><br>Python has several built-in data types. Each data type determines the operations that can be performed on the data and the way it is stored in memory.

| Data Type Category | Data Type               | Description                                   | Example                        |
|--------------------|-------------------------|-----------------------------------------------|--------------------------------|
| Numeric Types      | Integer (`int`)           | Whole numbers without a decimal point.        | `42, -1, 2024`                 |
|                    | Float (`float`)           | Numbers with a decimal point.                 | `3.14, -0.001`                 |
|                    | Complex (`complex`)       | Numbers with a real and an imaginary part.    | `2 + 3j`                       |
| Boolean Type       | Boolean (`bool`)          | Represents truth values.                      | `True, False`                  |
| Sequence Types     | List                    | Ordered and mutable collection of items.      | `[1, 'human', 3.14]`           |
|                    | Tuple                   | Ordered and immutable collection of items.    | `(1, 'human', 3.14)`           |
|                    | String (`str`)            | Sequence of characters.                       | `"Hello Detective"`, `'Thu'`              |
| Set Types          | Set                     | Unordered collection of unique items.         | `{1, 2, 3}`                    |
|                    | Frozen Set (`frozenset`)  | Immutable version of a set.                   | `frozenset([1, 2, 3])`         |
| Mapping Type       | Dictionary (`dict`)       | Collection of key-value pairs.                | `{'name': 'Alice', 'age': 25}` |
| Binary Types       | Bytes                   | Immutable sequence of bytes.                  | `b'Hello'`                     |
|                    | Bytearray               | Mutable sequence of bytes.                    | `bytearray([119, 105, 107, 105])` |

Both lists and dictionaries are both very useful data types. But they have the following key differences:

You found a secret message that the suspect sent to his associates: <br> 
"Phhw dw wkh fbehu fdih dw plgqljkw"

After some investigation, your team found that this message was encrypted using a simple shift cipher on the English alphabet.

Here's how it works:
<br><br>
<img src="images/shift_cipher.png" width="400"/>


<div style="background-color: #2e3440; padding: 8px; border-left: 4px solid #a3be8c; margin: 10px 0;">
<span style="color: #a3be8c;">üìä Task:</span> Use the function <span style="color: #81a1c1;">decode_encrypted_message()</span> below to decode the message.
</div>

</div>


```python
def decode_encrypted_message(message):
    """
    Decodes an encrypted message using a simple shift cipher.
    """
    decoded = ""
    for char in message:
        if char.isalpha():
            ascii_offset = 65 if char.isupper() else 97
            decoded += chr((ord(char) - ascii_offset - 3) % 26 + ascii_offset)
        else:
            decoded += char
    return decoded
```
